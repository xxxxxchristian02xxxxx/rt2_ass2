```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Image
import ipywidgets as widgets
import numpy as np
from io import BytesIO

# ---------------------------------------
# Dati globali
# ---------------------------------------
xdata, ydata = [], []
target_reached_list = []
target_cancelled_list = []

# Output widgets
out1 = widgets.Output()
out2 = widgets.Output()

# ---------------------------------------
# Funzione per convertire una figura in immagine
def fig_to_image(fig):
    buf = BytesIO()
    fig.savefig(buf, format='png')
    buf.seek(0)
    return Image(data=buf.read())

# ---------------------------------------
# Funzione per aggiornare grafico a torta
def update_pie_plot():
    fig, ax = plt.subplots()
    total = len(target_reached_list) + len(target_cancelled_list)
    if total == 0:
        sizes = [0.5, 0.5]
    else:
        sizes = [
            len(target_reached_list) / total,
            len(target_cancelled_list) / total
        ]
    ax.pie(sizes, labels=['Reached', 'Cancelled'], autopct='%1.1f%%')
    ax.set_title("Reached vs Cancelled")
    img = fig_to_image(fig)
    plt.close(fig)
    with out1:
        out1.clear_output(wait=True)
        display(img)

# ---------------------------------------
# Funzione per aggiornare grafico live
def update_live_plot():
    fig, ax = plt.subplots()
    ax.plot(xdata, ydata, 'r-')
    ax.set_xlim(max(0, xdata[-1] - 10) if xdata else 0, xdata[-1] + 0.1 if xdata else 10)
    ax.set_ylim(min(ydata) - 0.5 if ydata else -1.5, max(ydata) + 0.5 if ydata else 1.5)
    ax.set_title("Live trajectory")
    img = fig_to_image(fig)
    plt.close(fig)
    with out2:
        out2.clear_output(wait=True)
        display(img)

# ---------------------------------------
# Funzione generale per simulare update
def update_all(frame):
    t = frame * 0.1
    xdata.append(t)
    ydata.append(np.sin(t))

    if frame % 20 == 0:
        if np.random.rand() > 0.5:
            target_reached_list.append('target')
        else:
            target_cancelled_list.append('target')
        update_pie_plot()

    update_live_plot()

# ---------------------------------------
# Visualizza i due output affiancati
display(widgets.HBox([out1, out2]))

# ---------------------------------------
# Simula l'aggiornamento ogni 100 ms
from threading import Timer

frame_counter = [0]

def periodic_update():
    update_all(frame_counter[0])
    frame_counter[0] += 1
    Timer(0.1, periodic_update).start()

periodic_update()
```
