```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

# ---------------------------------------
# Dati globali
# ---------------------------------------
xdata, ydata = [], []
target_reached_list = []
target_cancelled_list = []

# ---------------------------------------
# Setup grafico a torta
# ---------------------------------------
fig1, ax1 = plt.subplots()
def update_pie():
    ax1.clear()
    total = len(target_reached_list) + len(target_cancelled_list)
    if total == 0:
        sizes = [0.5, 0.5]
    else:
        sizes = [len(target_reached_list) / total,
                 len(target_cancelled_list) / total]
    ax1.pie(sizes, labels=['Reached', 'Cancelled'], autopct='%1.1f%%')
    ax1.set_title("Reached vs Cancelled")
    fig1.canvas.draw()

update_pie()  # iniziale

# ---------------------------------------
# Setup grafico live
# ---------------------------------------
fig2, ax2 = plt.subplots()
ln, = ax2.plot([], [], 'r-')
ax2.set_xlim(0, 10)
ax2.set_ylim(-1.5, 1.5)
ax2.set_title("Live trajectory")

# ---------------------------------------
# Funzione di aggiornamento live
# ---------------------------------------
def update_plot(frame):
    t = frame * 0.1
    xdata.append(t)
    ydata.append(np.sin(t))

    ln.set_data(xdata, ydata)
    ax2.set_xlim(max(0, t - 10), t + 0.1)
    ax2.set_ylim(min(ydata) - 0.5, max(ydata) + 0.5)

    # Aggiorna grafico a torta ogni 20 frame
    if frame % 20 == 0:
        if np.random.rand() > 0.5:
            target_reached_list.append('target')
        else:
            target_cancelled_list.append('target')
        update_pie()

    return ln,

# ---------------------------------------
# Avvia animazione
# ---------------------------------------
ani = FuncAnimation(fig2, update_plot, interval=100)
```
