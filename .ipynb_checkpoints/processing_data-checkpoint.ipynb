```python
import yaml 
import numpy as np
with open("random_spawn_positions.yaml","r") as f:
    data = yaml.safe_load(f)
with open("results_targets1.yaml", "r") as f_s:
    data_result1_raw = list(yaml.safe_load_all(f_s))
    data_result1 = []
    for item in data_result1_raw:
        if isinstance(item, list):
            data_result1.extend(item)  # aggiunge tutti i dizionari della lista
        elif isinstance(item, dict):
            data_result1.append(item)  # aggiunge il singolo dizionario

with open("results_targets2.yaml","r") as f_s:
    data_result2_raw = list(yaml.safe_load_all(f_s))
    data_result2 = []
    for item in data_result2_raw:
        if isinstance(item, list):
            data_result2.extend(item)  # aggiunge tutti i dizionari della lista
        elif isinstance(item, dict):
            data_result2.append(item)  # aggiunge il singolo dizionario
```

```python
def convert_types(obj):
    if isinstance(obj, dict):
        new_dict = {}
        for k, v in obj.items():
            # Converti stringhe 'True'/'False' in bool
            if isinstance(v, str):
                if v.lower() == 'true':
#                     print("1")
                    new_dict[k] = True
                elif v.lower() == 'false':
#                     print("2")
                    new_dict[k] = False
                else:
                    try:
#                         print("3")
                        # prova a convertire in float
                        new_dict[k] = float(v)
                    except:
                        new_dict[k] = v
            else:
                new_dict[k] = convert_types(v)
        return new_dict
    elif isinstance(obj, list):
        return [convert_types(i) for i in obj]
    else:
        return obj
    
data_result1 = convert_types(data_result1)
data_result2 = convert_types(data_result2)
```

```python
def get_times(data):
    duration_vect = []
    terget_vect = []
    spawn_vect = []
    valid_vect = []
    connta = 0 
    for elem in data:
        nam = list(elem.keys())[0] 
#         if elem[nam]['validity'] == False:
#             connta += 1
#             continue
        duration_vect.append(elem[nam]['duration'])
        pos_ = (elem[nam]['target']['aim_x'],elem[nam]['target']['aim_y'])
        terget_vect.append(pos_)
        pos = (elem[nam]['spawn_position']['x'],elem[nam]['spawn_position']['y'])
        spawn_vect.append(pos)
        valid_vect.append(elem[nam]['validity'])
    print(connta)
    return duration_vect,terget_vect,spawn_vect,valid_vect

def get_times_prooned(data1, data2):
    duration_vect1 = []
    duration_vect2 = []

    for i, elem1 in enumerate(data1, start=1):
        # Ricava il nome del target per entrambi
        nam1 = list(elem1.keys())[0]
        elem2 = data2[i-1]
        nam2 = list(elem2.keys())[0]

        # Ottieni i due dizionari interni
        val1 = elem1[nam1]
        val2 = elem2[nam2]

        # Verifica la validità
        if not val1['validity'] or not val2['validity']:
            continue

        # Aggiungi le durate
        duration_vect1.append(val1['duration'])
        duration_vect2.append(val2['duration'])

    return duration_vect1, duration_vect2

```

```python
def mean_eval(data):
     return np.mean(data)
    
```

```python
times_p1 , times_p2=get_times_prooned(data_result1,data_result2)
```

```python


for elem in data_result1:
    key = list(elem.keys())[0]
#     print(elem[key]['duration'])

times1, targets1,spawns1,valid1 = get_times(data_result1)
times2, targets2,spawns2,valid2 = get_times(data_result2)


mean1 = np.mean(times_p1)
mean2 = np.mean(times_p2)
std1 = np.std(times_p1)
std2 = np.std(times_p2)



samples = np.array(times_p2) - np.array(times_p1)
n = len(samples)
mean_sam = np.mean(samples)
std_sam = np.std(samples)

std_error= std_sam/ np.sqrt(n)
t_statistical = mean_sam/std_error

# df = n -1


print(mean1)
print(mean2)
print(std1)
print(std2)
print(f"N = {n}")
print(f"mean diff :{mean_sam}")
print(f"std : {std_sam}")
print(f"t_statistical : {t_statistical}")

```

```python
vector = times1

# Dividi i dati per riempire prima la prima tabella
half = (len(vector) + 1) // 2  # arrotonda per eccesso
left_data = vector[:half]
right_data = vector[half:]


latex_code = r"""

\noindent
\begin{minipage}[t]{0.495\textwidth}
\centering
\begin{tabular}{|lllllll|}
\toprule
ID & Target & A1 & A2 & Diff & Valid1 & Valid2\\
\midrule
"""

# Riempie la prima tabella (a sinistra)
for i, value in enumerate(left_data, start=1):
    val_str = 'T' if valid1[i-1] else 'F'
    val_str2 = 'T' if valid2[i-1] else 'F'
    if val_str== 'F' or val_str2 == 'F' :
        continue
    latex_code += f"{i} & {targets1[i-1]} & {value:.2f} & {times2[i-1]:.2f} & {samples[i-1]:.2f} & {val_str} & {val_str2} \\\\\n"


latex_code += r"""\bottomrule
\end{tabular}
\captionof{table}{Tabella 1}
\end{minipage}
\hfill
\begin{minipage}[t]{0.495\textwidth}
\centering
\begin{tabular}{|lllllll|}
\toprule
ID & Target & A1 & A2 & Diff & Valid1 & Valid2 \\
\midrule
"""

# Riempie la seconda tabella (a destra)
for i, value in enumerate(right_data, start=half+1):
    idx = i - 1  # indice corretto in lista zero-based
    val_str = 'T' if valid1[idx] else 'F'
    val_str2 = 'T' if valid2[i-1] else 'F'
    if val_str== 'F' or val_str2 == 'F' :
        continue
    latex_code += f"{i} & {targets1[i-1]} & {value:.2f} & {times2[i-1]:.2f} & {samples[i-1]:.2f} & {val_str} & {val_str2} \\\\\n"


latex_code += r"""\bottomrule
\end{tabular}
\captionof{table}{Tabella 2}
\end{minipage}

"""

print(latex_code)
```
