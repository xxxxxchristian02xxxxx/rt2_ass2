```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import Button, Layout, HBox, VBox
from IPython.display import display,clear_output,HTML
import matplotlib.pyplot as plt

# Example data
sizes = [40, 60]
labels = ['Reached', 'Cancelled']

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.0f', startangle=90)

# Title
ax.set_title("Reached vs Cancelled Targets")

# Display the pie chart
plt.show()
a = 1
b = 2

text_x = widgets.Text(value = "0.0", layout = widgets.Layout(width = '80px'))
text_y = widgets.Text(value = "0.0",layout = widgets.Layout(width = '80px'))
text_x.value =str(a)
text_y.value = str(b)

text_x.disabled = True
text_y.disabled = True

# Titolo come HTML con margine per centrare verticalmente
label = widgets.HTML(
    value='<h2 style="margin:0; padding-right:52px;">Position (x,y):</h2>',
    layout=widgets.Layout(display='flex', align_items='center')
)

# Contenitore centrato
box = HBox(
    [label, text_x, text_y],
    layout=widgets.Layout(
        justify_content='flex-start', 
        align_items='center',      
        width='100%'               
    )
)

text = widgets.Text(value = "0.0",layout = widgets.Layout(width = '80px'))
text.disabled = True
label2 = widgets.HTML(
    value='<h2 style="margin:0px; padding-right:10px;">Closest obstacle:</h2>',
    layout=widgets.Layout(display='flex', align_items='center')
    
)

label3 = widgets.HTML(
    value='<h2 style="margin:0px; padding-left:10px;">metres</h2>',
    layout=widgets.Layout(display='flex', align_items='center')
)

box2 = HBox(
    [label2, text, label3],
    layout=widgets.Layout(justify_content='flex-start', align_items='center', width='100%')
)


display(box)
display(box2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random



random_xs = [random.randint(-9, 9) for _ in range(30)]

# Istogramma della distribuzione
plt.hist(random_xs, bins=100, alpha=0.7, label='Distribuzione Normale')
plt.title('Normal distribution with $\mu=0,\sigma=1$')
handles, labels = plt.gca().get_legend_handles_labels()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Intervallo possibile: da -9 a 9 inclusi (19 numeri totali)
possible_values = list(range(-9, 10))

# Prendi tutti i 19 valori unici (nessun duplicato)
random_xs = random.sample(range(-20, 21), 30)  # 30 valori unici tra -20 e 20

# Istogramma della distribuzione
plt.hist(random_xs, bins=19, alpha=0.7, label='Distribuzione Uniforme')
plt.title('Distribuzione uniforme di interi unici tra -9 e 9')
plt.legend()
plt.show()
```
