```python
import numpy as np, random
import matplotlib.pyplot as plt
import yaml


def get_random(num_departures, code):
    count = 0
    random_vector = []
    while count != num_departures:
        num_x = random.randint(-9, 9)
        num_y = random.randint(-9, 9)
        rand_pos = (num_x, num_y)
        if verify_target(rand_pos, random_vector) and code == 1 :
            random_vector.append(rand_pos)
            count += 1
        elif verify_target_telep(rand_pos, random_vector) and code == 2:
            random_vector.append(rand_pos)
            count += 1
    return random_vector

def verify_target(rand_pos, existing_positions):
    if rand_pos in existing_positions: 
        return False
    x_pos, y_pos = rand_pos 
    if  (
        (-5.5 <= x_pos <= 4 and -1 <= y_pos <= 0.5) or
        (-5.5 <= x_pos <= 4 and -5 <= y_pos <= 0.5) or
        (-7 <= x_pos <= 5 and 4 <= y_pos <= 5)

        ):
        return False
    return True


def verify_target_telep(rand_pos, existing_positions):
    if rand_pos in existing_positions: 
        return False
    x_pos, y_pos = rand_pos 
    if (
        (-5.5 <= x_pos <= 4 and -1 <= y_pos <= 0.5) or
        (-5.5 <= x_pos <= 4 and -5 <= y_pos <= 0.5) or
        (-7 <= x_pos <= 5 and 4 <= y_pos <= 5) or
        (x_pos >= 8 and y_pos >= 7) or
        not (-8 <= x_pos <= 8 and -8 <= y_pos <= 8)
    ):
        return False
    return True


num_departures = 10
random_positions_target = get_random(num_departures,1)
random_positions_spawn = get_random(num_departures*2,2)

# random_positions_target = [(3, -5),(-8, -7),(2,-3)]
# random_positions_target = [(-8, -7),(3, -5),(2,-3)]

target_random_list =random_positions_target
print(f'positions:{random_positions_target}')
random_spawn_position =random_positions_spawn
print(f'spawn:{random_positions_spawn}')

# fig, ax = plt.subplots()
# ax.hist(random_positions, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
# ax.set_title('Istogramma dei Dati (Truncated Normal)')
# ax.set_xlabel('Valori')
# ax.set_ylabel('Densità')
# plt.show()



with open("random_targets_generated.yaml", "w") as f:
    pass
    yaml.dump([list(pos) for pos in target_random_list], f)


with open("random_spawn_positions.yaml", "w") as f_p:
    pass
    yaml.dump([list(pos) for pos in random_spawn_position], f_p)
```
