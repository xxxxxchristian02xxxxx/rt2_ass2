```python
#!/usr/bin/env python3
import rospy
from actionlib import SimpleActionClient, TerminalState
from assignment_2_2024.msg import PlanningAction, PlanningGoal, PlanningFeedback
from nav_msgs.msg import Odometry
from geometry_msgs.msg import Pose, PoseStamped, Point,Twist
from std_msgs.msg import Float32
from sensor_msgs.msg import LaserScan
import time
import ipywidgets as widgets
from ipywidgets import Button, Layout, HBox, VBox
from IPython.display import display,clear_output
import matplotlib.pyplot as plt

#import for visualization
import tf
from tf.transformations import quaternion_matrix
import numpy as np
from matplotlib.animation import FuncAnimation

stop_toggle = True
vel :Twist = Twist()
xdata, ydata = [],[]


target_given_list =[]
target_reached_list =[]
target_cancelled_list= []

# variabes to deal with the goal definition
stat: str = ""
pose: Pose = Pose()
goal: PlanningGoal = PlanningGoal()
goal_past : PlanningGoal = PlanningGoal()
    
# lables to debug
lab_pos = widgets.Label(value = "(x: 0.0, y:0.0)")
lab_short_dis = widgets.Label(value =" : 0.0")
arrived_widget = widgets.Output() #quetsa forse on serve 
lab_feed = widgets.Label(value = "fdk")

# widget for the target vectors
lab_target_given = widgets.Label(value =f'Target given list: {target_given_list}')
lab_target_reached = widgets.Label(value =f'Target reached list: {target_reached_list}')
lab_target_cancelled = widgets.Label(value = f'Target reached list: {target_cancelled_list}')


# Generation of buttons
start_b = widgets.Button(description=' Start', layout=Layout(width='50%', height='80px'), disabled =False)
stop_b = widgets.Button(description='Stop', layout=Layout(width='50%', height='80px'),disabled =True)
cancel_b = widgets.Button(description='Cancel target', layout=Layout(width='50%', height='80px'),disabled =True)

given_list_output = widgets.Output()
reached_list_output = widgets.Output()
cancelled_list_output = widgets.Output()
```

# callbacks

```python
def pos_vel_callcback(msg):
    global vel , xdata, ydata
    vel = msg.twist.twist
    lab_pos.value = f'(x: {msg.pose.pose.position.x :.2f}, y:{msg.pose.pose.position.y :.2f})'
    xdata.append( msg.pose.pose.position.x)
    ydata.append( msg.pose.pose.position.y)
    
def feedback_callback(feedback_: PlanningFeedback) -> None:
    global stat
    stat = feedback_.feedback.stat
    lab_feed.value =f'feedback :{stat}'
    process_feedback()

def laser_callback(msg):
    global min_distance
    min_distance = float('inf')
    for range in msg.ranges:
        if range < min_distance:
            min_distance = range
    closest_obstacle_msg = Float32()
    closest_obstacle_msg.data = min_distance
    publish_distance.publish(closest_obstacle_msg)
    lab_short_dis.value = f' {min_distance:.2f}'
```

# buttons behaviours

```python
def on_start_button_clicked(b):
    with output_start:
        output_start.clear_output()
        print("started clicked.")
    start_b.disabled = True
    stop_b.disabled  = False
    cancel_b.disabled = False
    toggle_input()
    do_goal()


def on_stop_button_clicked(b):
    global stop_toggle, client, goal, goal_past, stop_b
    with output_stop:  
        output_stop.clear_output()
        print(f'valeu of toggle:{stop_toggle}')
    if(stop_toggle == True or stop_b.description == "Stop"):
        print(f'stop:valeu of toggle:{stop_toggle} (True) e valore è: {stop_b.description } ( Stop) ')
        stop_toggle = False
        stop_b.description ="Resume"
        goal_past = goal
        client.cancel_goal()

    elif(stop_toggle == False or stop_b.description == "Resume") :
        print(f'stop:valeu of toggle:{stop_toggle} (False) e valore è: {stop_b.description } (Resume) ')
        goal = goal_past
        client.send_goal(goal)
        stop_toggle = True
        stop_b.description = "Stop"
        
    
def on_cancel_button_clicked(b):
    global stop_b,stop_toggle
    with output_cancel:
        output_cancel.clear_output()
        print("cancel clicked.")
    cancel_b.disabled = True
    stop_b.disabled =True
    start_b.disabled =False
    toggle_input()
    x_input.value = 0.0
    y_input.value = 0.0

    print(f'cancel:valeu of toggle:{stop_toggle} (True) e valore è: {stop_b.description } ( Stop) ')

    del_goal()
    
def on_ini_button_clicked(b):
    start_b.disabled =False
    ini_b.disabled=True

def toggle_input():
    if( x_input.disabled == False and y_input.disabled == False ):
        x_input.disabled = True
        y_input.disabled = True
    elif(x_input.disabled == True and y_input.disabled == True):
        x_input.disabled = False
        y_input.disabled = False
```

# Functions for the robot 

```python
def process_feedback():
    global stat, target_reached_list, target_given_list,stop_toggle, stop_b
#     with arrived_widget:
#         arrived_widget.clear_output(wait=True)
#         print(stat)
           
    if(stat == "Target reached!"):
        target_reached_list.append(target_given_list[-1])  # put last element of the list since i do one at time
        with arrived_widget:
            arrived_widget.clear_output(wait=True)
            print("arrived")
        cancel_b.disabled = True
        stop_b.disabled =True
        start_b.disabled =False
        toggle_input()
        x_input.value = 0.0
        y_input.value = 0.0
        del_goal()
        print(f'process:valeu of toggle:{stop_toggle} (True) e valore è: {stop_b.description } ( Stop) ')

        lab_target_reached.value = f'Target reached list: {target_reached_list}'
        update_pie()
            
    if(stat == "Target cancelled!" and stop_toggle == True):
        print(f'process:valeu of toggle:{stop_toggle} (True) e valore è: {stop_b.description } ( Stop) ')
        del_goal()
        target_cancelled_list.append(target_given_list[-1])
        lab_target_cancelled.value= f'Target cancelled list: {target_cancelled_list}'
        update_pie()
    

    
def do_goal():
    global pose, goal, x_input,y_input, client, target_given_listù
    del_goal()
    pose.position.x = x_input.value 
    pose.position.y = y_input.value
    
    # keep counting the number of taets given and add to the list of targets given
    target = {'aim_x' :0.0,'aim_y' :0.0 }
    target['aim_x'] = pose.position.x 
    target['aim_y'] = pose.position.y
    target_given_list.append(target)
    lab_target_given.value = f'Target given list: {target_given_list}'
    goal.target_pose.pose = pose
    client.send_goal(goal) 


def del_goal():
    global client,stop_b,stop_toggle
    client.cancel_all_goals()
    if stop_b.description =="Resume":
        print(f'del:valeu of toggle:{stop_toggle} (False) e valore è: {stop_b.description } ( Resume) ')
        stop_toggle = True
        stop_b.description ="Stop"

```

# Interface

```python
rospy.init_node("target_client")
client: SimpleActionClient = SimpleActionClient("reaching_goal", PlanningAction)
fdk= rospy.Subscriber("/reaching_goal/feedback", PlanningFeedback, feedback_callback , queue_size=10)

rospy.Subscriber("/odom", Odometry, pos_vel_callcback)
pub_vel = rospy.Publisher("/cmd_vel", Twist,queue_size = 10 )
rospy.Subscriber('/scan', LaserScan, laser_callback)
publish_distance = rospy.Publisher('closest_obstacle', Float32, queue_size=10)
del_goal()

# Input boxes for target
message_label = widgets.Label("Goal taget:")
x_input = widgets.FloatText(description="X Target", value=0.0)
y_input = widgets.FloatText(description="Y Target", value=0.0)


# Generation of buttons
start_b = widgets.Button(description=' Start', layout=Layout(width='50%', height='80px'), disabled =False)
stop_b = widgets.Button(description='Stop', layout=Layout(width='50%', height='80px'),disabled =True)
cancel_b = widgets.Button(description='Cancel target', layout=Layout(width='50%', height='80px'),disabled =True)


# Output widget
output_start = widgets.Output()
output_stop = widgets.Output()
output_cancel = widgets.Output()


# binding buttons and functions
start_b.on_click(on_start_button_clicked)
stop_b.on_click(on_stop_button_clicked)
cancel_b.on_click(on_cancel_button_clicked)



# Widget visualization
display(widgets.Label("Set a target"))
display(HBox([x_input, y_input]))
display(widgets.Label("Choose operation"))
display(HBox([start_b, stop_b, cancel_b]))
display(HBox([output_start,  output_cancel]))
display(HBox([widgets.Label("Closer obsatcle at meters:"),lab_short_dis]))
display(HBox([widgets.Label("Position: "),lab_pos]))
display(lab_feed)
display(arrived_widget)
display(lab_target_given)
display(lab_target_reached)
display(lab_target_cancelled)
display(output_stop)
```
